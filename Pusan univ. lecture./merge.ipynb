```python
df1 = pd.concat([df,df_area],axis=1)   
df1
#단순히 열로 합쳤다.  던순히 concat는 key를 인덱스기준으로 dataframe을 붙혀준다. 만약 인덱스가 구명으로 같다면 결과는 merge와 같을 것이다.
```

```python
df2 = pd.merge(df, df_area, on='구명', how='inner', sort=False)
```

```python
df2
```

```python
import folium
```

```python
map = folium.Map(location=[35.150836, 129.053237],zoom_start=11)  


for i in df2.index :
    lat = df2.loc[i,'위도']  
    lang = df2.loc[i,'경도']
    
    
    folium.CircleMarker([lat,lang],     
                        radius =df2.loc[i,'전기'],
                        popup = df2.loc[i,'전기'],   
                        color = 'red',
                        fill = True).add_to(map)
    folium.CircleMarker([lat,lang],     
                        radius =df2.loc[i,'가스'],
                        popup = df2.loc[i,'가스'],   
                        color = 'blue',
                        fill = True).add_to(map)
    folium.CircleMarker([lat,lang],     
                        radius =df2.loc[i,'지역난방'],
                        popup = df2.loc[i,'지역난방'],   
                        color = 'green',
                        fill = True).add_to(map)
map    
```
