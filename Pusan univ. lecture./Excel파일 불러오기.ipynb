```python
busan = pd.read_excel('./부산구청.xlsx')
busan = busan.set_index("항목")
busan.loc['중구']
```

```python
for i in busan.index :
    lat = busan.loc[i,'위도']     #각각에 해당하는 위치정보를 가져옴
    lang = busan.loc[i,'경도']
    
    i = i.strip()
    print(lat,',',lang,',',i)   
    
# 각각의 위도 경도를 가져올 수 있다.
```
