```python
ap2 = ap.dropna(axis=1  # 열로 봤을 때, NA값이 하나라도 있는 열은 지움.
```

```python
ap2 = ap.dropna(thresh=44, axis=1)   # NA가 44이상인 열들을 지우겠다.
```

```python
ap2
```

```python
ap2.columns = ['시','구']
ap2.head()
```

```python
ap2['구'] = ap2['구'].fillna('*')   # NA에 *로 채움.
```

```python
ap2
```

```python
ap2['지역'] = ap2.apply(lambda x : x.시 if x.구 == '*' or x.구 == '' else x.구, axis=1)   #axis='columns'  
# 기본적으로 행으로 한줄씩 읽어오기 때문에 2개 씩 들어오게 된다. 옵션을 컬럼별로 한다면 처리가능.
```

```python
ap2
```

```python
ap2g = ap2.groupby('지역')['시'].count()      #ap2g = ap2.groupby('지역')[['시']].count()로 바로 dataframe으로 처리해도 된다.
```

```python
ap2g
```

```python
type(ap2g)
```

```python
ap2g = pd.DataFrame(ap2g)
```

```python
type(ap2g)
```

```python
ap2g
```

```python
ap2g.plot()
plt.show()
```

```python
ax = ap2g.plot(kind='bar',figsize=(10,6))

for i in ax.patches :     # 사각형 박스 하나씩을 들고온다.
    l, bt, w, h =  i.get_bbox().bounds   # 왼쪽 밑 오른쪽 높이에 대한 정보 4개를 반환한다.
    ax.annotate(h, (l+ w/2, h*1.05), ha='center' )                       # 무엇을 찍을 건지 할당. 높이를 찍고 싶음

    
    
    
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('지역별 인원수')    
plt.show()
```
