중심극한정리를 파이썬으로  

```python
import numpy as np
np.random.uniform(low=0, high=1, size=10)
```

```python
import matplotlib.pyplot as plt
import math
import matplotlib.mlab as mlab
import scipy.stats as ss
```

```python
%matplotlib inline
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu-3*sigma,mu+3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma), 'y', label="standard normal")      #표준정규분포 그래프의 모형
```

```python
#표본이 2개일때

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu-3*sigma,mu+3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma), 'y', label="standard normal")

i=()
for x in range(1,10000+1):
    k=np.random.uniform(low=0, high=1, size=2).mean()
    i+=(k,)
z=ss.zscore(i)    #표준화를 시켜줌
plt.hist(z,bins="auto", normed=True, facecolor = "none", edgecolor="black")
```
